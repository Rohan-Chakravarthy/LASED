# Calcium Simulation

Calcium ground state to its first excited state is one of the simplest systems to model as there is only a ground S-state and an excited P-state with no hyperfine structure. We can verify results already published by [Murray 2003](https://iopscience.iop.org/article/10.1088/0953-4075/36/24/011/meta) by using the `LASED` package.

First, we'll set up the states. Calcium's state setup is seen below:

Now, we'll import the `LASED` library and a plotting library. I use `Plotly` here but any plotting library can be used.

```python
%run "base-LASED/LASED/laser_atom_system.py"
%run "base-LASED/LASED/time_evolution.py"
%run "base-LASED/LASED/generate_sub_states.py"

import plotly.graph_objects as go
import time
```

Declare the system's variables: wavelength, lifetime, isospin, selection rules etc.

```python
wavelength_ca = 422.8e-9  # in m
w_e = te.angularFreq(wavelength_ca)

tau_ca = 4.6 # in ns
I_ca = 0  # Isospin of calcium
Q_decay = [1, 0, -1]
```

Create the state objects

```python
# Create states
s1 = State(label = 1, w = 0, m = 0, L = 0, S = 0)
s2 = State(label = 2, w = w_e, m = -1, L = 1, S = 0)
s3 = State(label = 3, w = w_e, m = 0, L = 1, S = 0)
s4 = State(label = 4, w = w_e, m = 1, L = 1, S = 0)
print(s3)

# Create ground and excited states list
G_ca = [s1]
E_ca = [s2, s3, s4]
```

Declare the laser parameters: intensity/power of the laser, polarisation of the laser, etc.

```python
intensity_ca = 100 # mW/mm^-2
Q_ca = [0]
detuning = 0.2*2*PI  # detuning here is 200 MHz in Grad/s
detuning2 = 0.5*2*PI  # detuning here is 500 MHz in Grad/s
```

Set the simulation parameters: list of time steps, etc.

```python
# Simulation parameters
start_time = 0
stop_time = 50 # in ns
time_steps = 501
time_ca = np.linspace(start_time, stop_time, time_steps)
```

Create a `LaserAtomSystem` using the variables stated above and perform a `timeEvolution`

```python
calcium_system = LaserAtomSystem(E_ca, G_ca, tau_ca, Q_ca, wavelength_ca, 
                                 laser_intensity = intensity_ca)

calcium_system200MHzdetuned = LaserAtomSystem(E_ca, G_ca, tau_ca, Q_ca, wavelength_ca, 
                                 laser_intensity = intensity_ca)
calcium_system500MHzdetuned = LaserAtomSystem(E_ca, G_ca, tau_ca, Q_ca, wavelength_ca, 
                                 laser_intensity = intensity_ca)

tic = time.perf_counter()
calcium_system.timeEvolution(time_ca,
                             pretty_print_eq = True)
calcium_system200MHzdetuned.timeEvolution(time_ca,
                             detuning = detuning)
calcium_system500MHzdetuned.timeEvolution(time_ca,
                             detuning = detuning2)
toc = time.perf_counter()
print(f"The code finished in {toc-tic:0.4f} seconds")
```

Now, we can plot the evolution of Calcium with no detuning, 200 MHz detuning, and 500 MHz detuning.

```python
rho_33 = [abs(rho) for rho in calcium_system.Rho_t(s3, s3)]
rho_33_200MHzdetuned = [abs(rho) for rho in calcium_system200MHzdetuned.Rho_t(s3, s3)]
rho_33_500MHzdetuned = [abs(rho) for rho in calcium_system500MHzdetuned.Rho_t(s3, s3)]

fig_ca = go.Figure()
fig_ca.add_trace(go.Scatter(x = time_ca,
                                 y = rho_33,
                                 mode = 'lines',
                                    name = "Rho_33 (δ = 0 MHz)",
                                    marker = dict(
                                color = 'red',
                                symbol = 'circle',
                            )))
fig_ca.add_trace(go.Scatter(x = time_ca,
                                 y = rho_33_200MHzdetuned,
                                 mode = 'lines',
                                    name = "Rho_33 (δ = 200 MHz)",
                                    marker = dict(
                                color = 'blue',
                                symbol = 'circle',
                            )))
fig_ca.add_trace(go.Scatter(x = time_ca,
                                 y = rho_33_500MHzdetuned,
                                 mode = 'lines',
                                    name = "Rho_33 (δ = 500 MHz)",
                                    marker = dict(
                                color = 'green',
                                symbol = 'circle',
                            )))

fig_ca.update_layout(title = "Calcium Upper State Population after Linear Excitation: I = 100 mW",
                 xaxis_title = "Time (ns)",
                 yaxis_title = "Population",
                font = dict(
                    size = 11))
fig_ca.show()
```
